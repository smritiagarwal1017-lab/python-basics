**Quest 2-** What is an interpreter in Python?

**Ans 2-** An interpreter in Python is a program that directly reads and executes Python source code. It acts as both a translator and a runtime environment, processing instructions line by line and converting them into machine-readable form so the computer can run them immediately, rather than compiling the entire program beforehand.

**Quest 3-** What are pre-defined keywords in Python?

**Ans 3-** Pre-defined keywords in Python are special reserved words with specific meanings and purposes that define the language's syntax and structure. They cannot be used as identifiers, such as variable, function, or class names. eg., if-else, for, while, if-elif-else, continue and break.

**Quest 4-** Can keywords be used as variable names?

**Ans 4-** No, keywords cannot be used as variable names in Python.

**Quest 5-** What is mutability in Python?

**Ans 5-** Mutability in Python refers to the ability of an object to be modified after it has been created. This means you can change its internal state or contents without creating a new object in memory.

**Quest 6-** Why are lists mutable, but tuples are immutable?

**Ans 6-** Lists are mutable to allow for dynamic data manipulation (adding/removing/changing elements), while tuples are immutable to provide fixed, constant, and safe data structures. Lists are ideal for ordered, changeable collections, whereas tuples represent rigid structures.

**Quest 7-** What is the difference between “==” and “is” operators in Python?

**Ans 7-** In Python, the **==** operator checks for value equality, while the **is** operator checks for object identity.

**Quest 8-** What are logical operators in Python?

**Ans 8-** Python has three logical operators: **and, or, and not**. These operators combine conditional statements and return a Boolean value (True or False), which is used to control the flow of a program based on multiple conditions.

**Quest 9-** What is type casting in Python?

**Ans 9-** Type casting, also known as type conversion, is the process of converting a value from one data type to another in Python. This is essential for ensuring data compatibility, performing operations that require specific data types (e.g., mathematical calculations), and handling
user input correctly. Python supports two main types of casting and they are implicit type casting and explicit type casting.

**Quest 10-** What is the difference between implicit and explicit type casting?

**Ans 10-**  In **Implicit Type Casting** python automatically converts one data type to another to prevent data loss or errors and in **Explicit Type Casting** the programmer manually converts one data type into another.

**Quest 11-** What is the purpose of conditional statements in Python?

**Ans 11-** Conditional statements in Python are used to control the flow of a program by making decisions, executing specific blocks of code only if certain conditions are met. They allow a program to behave differently in various situations, responding dynamically to data and user input.  By using conditional logic and functions, you can avoid copying and pasting the same code blocks repeatedly, making the code easier to update and maintain.

**Quest 12-** How does the elif statement work?

**Ans 12-** The elif statement in Python (short for "else if") is used to check multiple conditions sequentially after an initial if statement. It allows a program to make a decision among several alternatives, executing the code block associated with the first condition that evaluates to True.

**Quest 13-** What is the difference between for and while loops?

**Ans 13-** **For loops** are best used when the number of iterations is known in advance, acting as a concise way to iterate over sequences or ranges. Conversely, **while loops** are used when the number of iterations is not known beforehand, relying on a specific boolean condition to continue.

Syntax (Example):

**For**: for (int i = 0; i < 10; i++) { ... }

**While**: while (condition) { ... }

**Quest 14-** Describe a scenario where a while loop is more suitable than a for loop.

**Ans 14-** A while loop is generally more suitable than a for loop in Python when the number of iterations is not known beforehand, and the loop needs to continue as long as a certain condition is met.

```python
# Quest 14- Write a program to reverse a string using a while loop.
# Answer
text=("apple")
index=len(text)-1
while index>=0:
  text+=text[index]
  index-=1
  print(text)
```

```python
# Quest 15- Write a Python program to calculate the factorial of a number provided by the user using a while loop.
# Answer
num=int(input("enter number"))
f=1
i=1
while i<=num:
  f*=i
  i+=1
  print(f)
```
